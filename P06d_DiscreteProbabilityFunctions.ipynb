# Discrete Probability Functions

We have four primary discrete probability functions that cover the majority of problem solving scenarios in a college probability course.

- <span style="color:red"><b>Binomial Distriubtion:</b></span> Prototypical coin flip distribution, coin can be fair or unfairly weighted. Also used for any situation that can summarized as a success/failure outcome with fixed probability.
- <span style="color:red"><b>Hypergeometric Distriubtion:</b></span> Dealing a hand of poker or other card game. Works well with any probability experiement that partitions easily into subsets.
- <span style="color:red"><b>Geometric Distriubtion:</b></span> Fixed probability where the probability experiment continues until a single success is achieved.
- <span style="color:red"><b>Negative Binomial Distriubtion:</b></span> Fixed probability where the probability experiment continues until a fixed number of $k$ successes have been achieved.

## Commands

We need to use the **choose()** function for combinations to create a function for permutations.

## Example Questions

First, determine the probability scenario described as binomial, hypergeometric, geometric, or negative binomial. Find the probability function for the following probability experiments, and answer the indicated question.

1. **Face Cards.** A standard deck has 52 cards, 12 of which are face cards. If you are dealt 5 cards, what is the probability that exactly 2 of them are face cards?

2. **The Job Interview.** A candidate has a 20% chance of receiving a job offer after any given interview. What is the probability that they receive their first offer on exactly their 4th interview?

3. **The "At Most" Scenario.** A seedsman claims that 90% of his seeds germinate. If you plant 6 seeds, what is the probability that at most 4 germinate?

4. **The Free Throw Streak.** A basketball player has a 70% free-throw average. What is the probability that they make their 5th basket on their 8th attempt?

5. **Software Bugs.** A developer is testing code. Each run has a probability $p = 0.4$ of triggering a specific rare bug. Find the probability that the bug occurs for the first time on an even-numbered run (i.e., $X \in \{2, 4, 6, \dots\}$).

6. **Candy Machines.** A vintage gumball machine contains 20% red gumballs. A child wants to get 2 red gumballs. What is the probability that the child will have to buy exactly 7 gumballs to get their 2nd red one?

7. **Blood Donation.** Only 7% of the population has Type O-negative blood. A blood bank is looking for a donor with this type. What is the probability that they will need to test more than 10 people to find the first O-negative donor?

8. **Professional Certification.** An exam bank contains 100 questions: 60 are "easy" and 40 are "hard." The computer randomly generates a 10-question test for a student. What is the probability that the student gets exactly 7 easy questions?

9. **The Software Bug.** In a beta test, a specific software interaction crashes 15% of the time. If the interaction is performed 5 times, find the entire probability distribution (calculate $P(X=k)$ for $k=0$ through $5$) and verify they sum to 1.

10. **Wildlife Tagging (Capture-Recapture).** Ecologists capture and tag 20 turtles in a pond and release them. A week later, they capture 15 turtles, and find that 5 of them are tagged. If the pond actually contains 50 turtles, what was the probability of getting exactly 5 tagged turtles in the second sample?

11. **Passing the Exam.** A student is taking a multi-part certification. They can retake the exam as many times as needed, but they must pass 4 modules to get certified. If the probability of passing any single module attempt is 0.6, find the probability they finish the certification on their 6th attempt.

12. **The Software Bug.** A programmer is running automated tests. Each test has a 20% chance of triggering a specific rare bug. What is the probability that the bug occurs for the 4th time on the 15th test run?


14. **Online Dating.** Suppose the probability of a "match" on a swipe-based app is 0.15. Let $X$ be the number of swipes until the first match. Calculate the variance of $X$.

15. **Jury Selection.** A jury pool of 25 people contains 10 people over the age of 50 and 15 people under 50. If a jury of 12 is selected at random, what is the probability that at least 8 of them are under the age of 50?

16. **The Free Throw Specialist.** A basketball player makes 80% of their free throws. In a series of 8 shots, what is the probability they miss exactly 3 times? (Careful with your $p$ value here!)

17. **The Persistent Salesman.** A door-to-door salesman has a 10% success rate for making a sale. If he visits doors until he makes his first sale, what is the probability that he has to visit at least 6 doors?

18. **Basketball Practice.** A player makes free throws with 80% accuracy. They decide to keep shooting until they miss for the first time. Note: Here, a "success" is defined as a miss. What is the probability they make exactly 5 shots before their first miss? (Meaning the miss happens on the 6th shot).


19. **Oil Drilling.** An oil company estimates that a well has a 25% chance of being productive. They need to find 3 productive wells to meet their quarterly goal. What is the probability they will need to drill exactly 10 wells to hit that goal?

20. **The Broken Lightbulbs.** A crate contains 50 lightbulbs. Unknown to the buyer, 5 of them are burnt out. If the buyertests 3 bulbs at random, what is the probability that the crate is accepted (meaning 0 bulbs are found to be burnt out)?

21. **Lottery Logic.** In a "6/49" lottery, 6 numbers are drawn from a set of 49. To win the second prize, you must match exactly 5 of the 6 winning numbers. What is the probability of winning the second prize with one ticket?

22. **Quality Control.** A factory produces lightbulbs with a 5% defect rate. An inspector tests bulbs one by one. What is the probability that the 2nd defective bulb is found on the 20th test?

23. **Network Packets.** In a noisy communication channel, a packet has a 0.1 probability of being corrupted. A receiver requires 5 uncorrupted packets to decode a message. What is the probability that the 5th clean packet is the 8th packet sent?

24. **The Board Game.** In a certain game, you need to roll a "6" on a fair six-sided die to move your piece out of the starting area. What is the expected number of rolls needed to start the game?

25. **The Quality Control Check.** A factory produces light bulbs with a 5% defect rate. If you randomly sample 10 bulbs, what is the probability that exactly 2 are defective?

26. **The Sales Call.** A telemarketer has a 12% chance of making a sale on any given call. Find the probability that they land their 3rd sale of the day on their 10th call.

27. **Committee Selection.** A department consists of 8 women and 6 men. A search committee of 4 people is to be formed. Find the probability that the committee consists of exactly 2 women and 2 men.

28. **Rare Pokémon Hunting.** In a specific area of a game, there is a 2% chance of encountering a "Shiny" Pokémon. What is the probability that a player encounters their 1st Shiny on exactly their 50th encounter? 29. **Quality Control.** A box contains 20 computer chips. 4 of the chips are defective. If a technician randomly selects 5 chips for testing, what is the probability that:
  - None are defective?
  - Exactly 1 is defective?

28. **Marketing Reach.** A digital ad has a click-through rate (CTR) of 2%. If the ad is shown to 50 people, what is the probability that more than 2 people click it?

29. **The Job Search.** An applicant has a 15% chance of receiving a job offer after an interview. If they are determined to keep interviewing until they receive 2 offers, what is the probability that their 2nd offer comes on their 5th interview?

30. **Quality Control.** In a manufacturing plant, 5% of the microchips produced are defective. An inspector tests chips one by one. What is the probability that the first defective chip is found within the first 3 tests?

31. **The Candy Jar.** A jar contains 15 red jellybeans and 10 blue jellybeans. If you grab a handful of 6 jellybeans at random, what is the probability that you have more red jellybeans than blue ones?

32. **The "Perfect" Guess.** A multiple-choice quiz has 15 questions, each with 4 options. If a student guesses randomly on every question, what is the probability they get exactly 10 correct?

33. **Garden Seeds.** A packet of seeds contains 12 sunflower seeds and 8 marigold seeds. You pick 5 seeds to plant in a small pot. What is the probability that the majority of the seeds (3 or more) are marigolds?

34. **The Rare Event.** A rare mutation occurs in 1% of a specific population. In a random group of 100 people, use the binomial PMF to find the probability that at least one person has the mutation. (Hint: Use the complement rule: $1 - P(X=0)$).
